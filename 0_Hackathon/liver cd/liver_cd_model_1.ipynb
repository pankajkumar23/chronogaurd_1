```python
# Importing Modules and Libraries
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Algorithm module
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
# Loading Data
columns = ['t_bilirubin', 'd_bilirubin', 'alkphos','sgpt', 'sgot', 'albumin', 'result']
df = pd.read_csv('/content/drive/MyDrive/0_Hackathon_SBSSU/liver cd/ilpd_md2.csv',usecols=columns)
df = df.astype({'alkphos':'int64','sgpt':'int64','sgot':'int64'})
df.head()
```

```python
# Train Test Split
max = -1
for i in range(1000):
  for j in range(1,5):
    k = j/10
    for l in range(2,6):
      x = df.drop('result',axis=1)
      y = df['result']

      random_state = i

      x_train, x_test, y_train, y_test = train_test_split(x,
                                                          y,
                                                          test_size=k,
                                                          random_state=random_state)

      # Model Training
      dt = DecisionTreeClassifier(max_depth=l)
      dt.fit(x_train,y_train)

      # y_pred = dt.predict(x_train)
      # accuracy = accuracy_score(y_train,y_pred)*100
      # print("Train accuracy : %.1f "%(accuracy))

      y_pred = dt.predict(x_test)
      accuracy = accuracy_score(y_test,y_pred)*100
      if accuracy > max:
        print("Test accuracy : %.1f "%(accuracy), " at {} i and {} k and L {}".format(i,j,l))
        max = accuracy
```
